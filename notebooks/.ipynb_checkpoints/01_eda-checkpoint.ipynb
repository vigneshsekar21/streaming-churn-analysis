# Exploratory Data Analysis - Streaming Subscriber Churn

This notebook explores churn patterns across subscription tier, engagement level, onboarding speed, and tenure window. After this, it builds a cohort retention curve to track subscriber retention over time.

**Key findings:**
- Engagement level is strongest predictor: low-engagement subscribers churn at 40.3% vs 6.3% for high-engagement.
- Onboarding speed (time to first watch) is a fast decline. Slow to first watch churn at 36% and faster watchers (under 1 week time) churn at 16%.
- Cohort retention curves track what percent of subscribers remained active at 1, 3, and 6 months after sign-up. This accounts for censoring out subscribers who have not yet met those time intervals. Recent cohorts that haven't had enough time to reach the checkpoint show NULL rather than misleading values.
- Number of titles watched shows a small but potentially relevant gradient. Low watchers churn at 22% whereas high watchers churn at 13%. This was not directly built into the churn formula but is likely correlated with engagement. Good candidate to test in the model if it adds independent predictive power or worth removing altogether.
- Genre/content is not a significant driver of churn. Each genre have uniform churn rates around 16-18%.
- Age group and region show no meaningful churn variation (16-18% range across both segments). consistent with region/age not being wired into the underlying churn generation logic. Worth explicitly ruling out for completeness, even though the result itself isn't surprising.

```python
import duckdb
import pandas as pd

# Load dataset
df = pd.read_csv("../streaming_subscribers.csv")

# register as a DuckDB table to write SQL against it
con = duckdb.connect()
con.register("subscribers", df)

# sanity check
# con.sql("SELECT * FROM subscribers LIMIT 5").show()
# con.sql("SELECT COUNT(*) FROM subscribers").show()
```

```python
# churn by tier
con.sql("""
SELECT 
subscription_tier,
COUNT(*) AS total_subscribers,
SUM(churned) AS churned_count,
ROUND(AVG(churned) * 100, 1) AS churn_rate_pct
FROM subscribers
GROUP BY subscription_tier
ORDER BY churn_rate_pct DESC
""").show()
# Basic users are roughly 6 times as likely to churn as premium users. Basic users churn at 30% while premium churns at 5%.
```

```python
# churn by engagement level
con.sql("""
    SELECT 
        CASE 
            WHEN avg_weekly_watch_hours < 2 THEN 'Low (<2 hrs)'
            WHEN avg_weekly_watch_hours < 6 THEN 'Medium (2-6 hrs)'
            ELSE 'High (6+ hrs)'
        END AS engagement_level,
        COUNT(*) AS total_subscribers,
        ROUND(AVG(churned) * 100, 1) AS churn_rate_pct
    FROM subscribers
    GROUP BY engagement_level
    ORDER BY churn_rate_pct DESC
""").show()
# Engagement level is the strongest predictor identified across all segments — low engagement users churn at 40% vs just 6.3% for high engagement, a 6x gap. The drop-off between low and medium/high watchers is sharp, suggesting engagement crosses a meaningful threshold rather than declining gradually.
```

```python
# churn by onboarding speed
con.sql("""
    SELECT
        CASE
            WHEN days_to_first_watch <= 3 THEN 'Fast starter (0-3 days)'
            WHEN days_to_first_watch <= 7 THEN 'Moderate (4-7 days)'
            ELSE 'Slow starter (8+ days)'
        END AS onboarding_speed,
        COUNT(*) AS total_subscribers,
        ROUND(AVG(churned) * 100, 1) AS churn_rate_pct
    FROM subscribers
    GROUP BY onboarding_speed
    ORDER BY churn_rate_pct DESC
""").show()
# This suggests the first 7 days post-signup is a critical intervention window — onboarding nudges or content recommendations during this period could meaningfully reduce churn.
```

```python
# churn by tenure window
con.sql("""
    SELECT
        CASE
            WHEN months_subscribed < 3 THEN 'New (<3 months)'
            WHEN months_subscribed < 12 THEN 'Established (3-12 months)'
            ELSE 'Loyal (12+ months)'
        END AS tenure_segment,
        COUNT(*) AS total_subscribers,
        ROUND(AVG(churned) * 100, 1) AS churn_rate_pct
    FROM subscribers
    GROUP BY tenure_segment
    ORDER BY churn_rate_pct DESC
""").show()
# Tenure alone is a weaker predictor than engagement. 
# In other words, a subscriber's behavior matters more than how long they've been a customer. 
# This means retention efforts should prioritize re-engaging low-activity users regardless of tenure, rather than assuming risk decreases simply with time.


```

```python
# Churn by age group, churn by region, and churn by titles watched
print("=== Churn by Age Group ===")
con.sql("""
    SELECT age_group, COUNT(*) AS total, ROUND(AVG(churned)*100, 1) AS churn_rate_pct
    FROM subscribers 
    GROUP BY age_group 
    ORDER BY churn_rate_pct DESC
""").show()

print("=== Churn by Region ===")
con.sql("""
    SELECT region, COUNT(*) AS total, ROUND(AVG(churned)*100, 1) AS churn_rate_pct
    FROM subscribers
    GROUP BY region
    ORDER BY churn_rate_pct DESC
""").show()

print("=== Churn by Titles Watched ===")
con.sql("""
    SELECT CASE
                WHEN num_titles_watched < 10 THEN 'Low (<10 titles)'
                WHEN num_titles_watched < 30 THEN 'Medium (10-30 titles)'
                ELSE 'High (30+ titles)'
            END AS titles_bucket,
            COUNT(*) AS total,
            ROUND(AVG(churned)*100, 1) AS churn_rate_pct
    FROM subscribers
    GROUP BY titles_bucket
    ORDER BY churn_rate_pct DESC
""").show()

# Ranking these findings by strength:
## Engagement level (6x gap)
## Tier (5x gap)
    ### While subscription tier shows a strong raw correlation with churn (5.35x ratio between Basic and Premium), 
    ### this is likely partially explained by underlying engagement differences across tiers rather than tier itself being a direct cause. 
    ### Engagement level remains the most actionable and direct predictor.
## Onboarding speed (2.3x gap)
## Tenure (1.7x gap)
## Number of titles watched (1.6x gap)
## Age group, Region
```

```python
# churn by genre/content
## Answers the question if there are content gaps that correlate with subscriber drop-off?
con.sql("""
    SELECT 
        favorite_genre,
        COUNT(*) AS total_subscribers,
        ROUND(AVG(churned) * 100, 1) AS churn_rate_pct
    FROM subscribers
    GROUP BY favorite_genre
    ORDER BY churn_rate_pct DESC
""").show()
# From this output, churn rate is nearly uniform across all favorite genres (16-18%).
# No genre stands out significantly. This suggests churn in this dataset is driven
# primarily by behavioral factors (engagement, onboarding speed) rather than content/genre preference.

## This directly answers one of the core business questions: content gaps do not appear 
## to be a meaningful churn driver. Note this is consistent with the dataset's design — 
## genre wasn't built as a churn factor in the underlying data generation logic, but in a 
## real-world setting, this kind of null result would still be valuable: it tells a content 
## team to stop looking at genre breadth and focus retention efforts on product engagement instead.
```

```python
# query to build cohort retention curve. 
# of all subscribers who signed up in a given month, what % are still active after 1, 3, 6 months?
# included observation_date field to ensure that subscribers who have not been subscribed long enough
# do not get included in the counts. For example, a subscriber from december 10, 2024, will have a NULL
# retained_month_1 field. This concept is called censoring.
con.sql("""
    WITH cohort_base AS (
        SELECT 
            subscriber_id,
            DATE_TRUNC('month', signup_date::DATE) AS signup_month,
            churned,
            churn_date::DATE AS churn_date,
            DATE '2024-12-31' AS observation_date
        FROM subscribers
    )
    SELECT 
        signup_month,
        COUNT(*) AS cohort_size,

        CASE WHEN observation_date >= signup_month + INTERVAL 1 MONTH THEN
            ROUND(AVG(CASE WHEN churned = 0 THEN 1
                            WHEN churn_date > signup_month + INTERVAL 1 MONTH THEN 1
                            ELSE 0 END) * 100, 1)
        ELSE NULL END AS retained_month_1,

        CASE WHEN observation_date >= signup_month + INTERVAL 3 MONTH THEN
            ROUND(AVG(CASE WHEN churned = 0 THEN 1
                            WHEN churn_date > signup_month + INTERVAL 3 MONTH THEN 1
                            ELSE 0 END) * 100, 1)
        ELSE NULL END AS retained_month_3,

        CASE WHEN observation_date >= signup_month + INTERVAL 6 MONTH THEN
            ROUND(AVG(CASE WHEN churned = 0 THEN 1
                            WHEN churn_date > signup_month + INTERVAL 6 MONTH THEN 1
                            ELSE 0 END) * 100, 1)
        ELSE NULL END AS retained_month_6

    FROM cohort_base
    GROUP BY signup_month, observation_date
    ORDER BY signup_month
""").show()
## Notice that 6-month retention for June 2024 joiners (80%) is meaningfully lower than 
## every cohort from 2023 (consistently >90%). Since this is synthetic/randomly generated 
## data, this is most likely natural sampling variation in a smaller cohort rather than a 
## real trend. In a production setting, this kind of dip would warrant investigation. 
## e.g. checking for a product change, pricing shift, or content gap introduced around 
## that signup period.
```

```python
# visualization of retention curve (use in Tableau dashboard)
import matplotlib.pyplot as plt

retention_df = con.sql("""
    WITH cohort_base AS (
        SELECT 
            subscriber_id,
            DATE_TRUNC('month', signup_date::DATE) AS signup_month,
            churned,
            churn_date::DATE AS churn_date,
            DATE '2024-12-31' AS observation_date
        FROM subscribers
    )
    SELECT 
        signup_month,
        COUNT(*) AS cohort_size,

        CASE WHEN observation_date >= signup_month + INTERVAL 1 MONTH THEN
            ROUND(AVG(CASE WHEN churned = 0 THEN 1
                            WHEN churn_date > signup_month + INTERVAL 1 MONTH THEN 1
                            ELSE 0 END) * 100, 1)
        ELSE NULL END AS retained_month_1,

        CASE WHEN observation_date >= signup_month + INTERVAL 3 MONTH THEN
            ROUND(AVG(CASE WHEN churned = 0 THEN 1
                            WHEN churn_date > signup_month + INTERVAL 3 MONTH THEN 1
                            ELSE 0 END) * 100, 1)
        ELSE NULL END AS retained_month_3,

        CASE WHEN observation_date >= signup_month + INTERVAL 6 MONTH THEN
            ROUND(AVG(CASE WHEN churned = 0 THEN 1
                            WHEN churn_date > signup_month + INTERVAL 6 MONTH THEN 1
                            ELSE 0 END) * 100, 1)
        ELSE NULL END AS retained_month_6

    FROM cohort_base
    GROUP BY signup_month, observation_date
    ORDER BY signup_month
""").df()

plt.figure(figsize=(10,5))
plt.plot(retention_df['signup_month'], retention_df['retained_month_1'], label='Month 1', marker='o')
plt.plot(retention_df['signup_month'], retention_df['retained_month_3'], label='Month 3', marker='o')
plt.plot(retention_df['signup_month'], retention_df['retained_month_6'], label='Month 6', marker='o')
plt.xlabel('Signup Month')
plt.ylabel('Retention %')
plt.title('Cohort Retention Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Sanity checked that the graph is correctly generated. Month 1 line never dips beneath month 3 or 6. 
# That would be impossible. 

## The gap between month 1 and month 6 retention widens for more recent cohorts. 
## This is most likely sample size noise — recent cohorts have smaller effective sample 
## sizes for the longer-window checkpoints, since fewer of their members have had enough 
## time to reach month 6. In a real-world setting, I would investigate whether this widening 
## gap holds up with statistical significance testing (e.g. confidence intervals on each 
## cohort's retention rate) before concluding it reflects an actual worsening trend, rather 
## than treating it as a confirmed finding.
```
